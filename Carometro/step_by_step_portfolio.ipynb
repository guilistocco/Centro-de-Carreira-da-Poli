```python

#bibliotecas de python-pptx
from pptx import Presentation
from pptx.util import Inches, Pt, Cm
from pptx.text.text import Font as font
from pptx.dml.color import RGBColor

# pandas para organização e carregamento dos dados
import pandas as pd
import numpy as np

# OS para abrirmos a apresentação assim que terminar
import os

#
from pydrive.auth import GoogleAuth
from pydrive.drive import GoogleDrive
```

## Implementacao com dados reais

```python


arquivo_excel = r'Compartilhamento de Dados de CV e Foto - cruzamento.csv' #ok
arquivo_ppt_template = r'AdP_Template PPT 2020 - Portifolio de alunos_exp.pptx' #ok
titulo_programa = "Trilha de Desenvolvimento - Processos Seletivos 2021" #ok
```

```python
arquivo_excel = r'Compartilhamento de Dados de CV e Foto.xlsx' #ok

cruzamento = pd.read_excel(arquivo_excel, sheet_name='cruzamento',
                     header=0,usecols=[0,1,2,3,4,5,6,7,8],nrows=101,
                     parse_dates=True, engine='openpyxl').sort_values('nome').reset_index(drop=True)

CV_lista = list(cruzamento['cv'].values)
CV_ids = []

for f in CV_lista:

    try:
        CV_ids.append(f[33:])
    except:
        CV_ids.append('nan')

cruzamento['cv_id'] = CV_ids


fotos_id = list(cruzamento['foto'].values)

f_ids = []

for f in fotos_id:
    try:
        f_ids.append(f[33:])

    except:
        f_ids.append('nan')

cruzamento['foto_id'] = f_ids

cruzamento
```

```python

#quando colocamos um argumento, ele busca por esse pptx no diretorio
prs = Presentation('AdP_Template PPT 2020 - Portifolio de alunos_exp.pptx')


# cria um slide com o layout 0 das opcoes, nesse template é o com titulo e subtitulo
title_slide_layout = prs.slide_layouts[1]
slide = prs.slides.add_slide(title_slide_layout)
title = slide.shapes.title
subtitle = slide.placeholders[0]

# coloca o titulo e subtitulo que queremos
title.text = "Portfólio de Alunos"
subtitle.text = titulo_programa

# Directory 
directory_fotos = "Fotos formulario"
directory_CV = "Resume Book"
  
# Diretorio atual
parent_dir = os.getcwd()

# Path 
caminho_fotos = os.path.join(parent_dir, directory_fotos) 
caminho_CV = os.path.join(parent_dir, directory_CV) 

if  os.path.isdir(caminho_fotos) == False:
    os.mkdir(caminho_fotos) 

if  os.path.isdir(caminho_CV) == False:
    os.mkdir(caminho_CV) 

```

```python

gauth = GoogleAuth()
gauth.LocalWebserverAuth()
drive = GoogleDrive(gauth)
```

```python
%%time

for row in cruzamento.index:
    
    # cria um slide com o layout 2 das opcoes, nesse template eh o com titulo e texto
    texto_slide_layout = prs.slide_layouts[6]
    slide = prs.slides.add_slide(texto_slide_layout)

    #coloca o nome como título
    title = slide.shapes.title
    # title.text = str(cruzamento["Nome completo"][row]) ### original
    title.text = str(cruzamento["nome"][row])

    # cria caixa de texto add_textbox(left, top, width, height)
    txBox = slide.shapes.add_textbox(Cm(16),Cm(5),Cm(17),Cm(13))
    tf = txBox.text_frame
    tf.word_wrap = True

# ###################
    #coloca email e outras informações
    #------------------- EMAIL -----------------
    p = tf.add_paragraph()
    p = tf.paragraphs[1]
    run = p.add_run()
    font = run.font
    font.color.rgb = RGBColor(46,188,218)

    font.bold = True
    run.text = "Email: " 
    run = p.add_run()
    font = run.font
    font.color.rgb = RGBColor(46,188,218)
    run.text = str(cruzamento['email'][row])

#     #------------------- TELEFONE -----------------

#     p = tf.add_paragraph()
#     p = tf.paragraphs[2]
#     run = p.add_run()
#     font = run.font
#     font.bold = True
#     run.text = "Telefone: " 
#     run = p.add_run()
#     run.text = str(cruzamento['Telefone'][row])

#     #------------------- CURSO -----------------


    p = tf.add_paragraph()
    p = tf.paragraphs[2]
    run = p.add_run()
    font = run.font
    font.color.rgb = RGBColor(46,188,218)

    font.bold = True
    run.text = "Curso: "
    run = p.add_run()
    font = run.font
    font.color.rgb = RGBColor(46,188,218)
    run.text = str(cruzamento['curso'][row])
    
#     #------------------- FORMATURA -----------------

    p = tf.add_paragraph()
    p = tf.paragraphs[3]
    run = p.add_run()
    font = run.font
    font.color.rgb = RGBColor(46,188,218)

    font.bold = True
    run.text = "Ano de Formatura: "
    run = p.add_run()
    font = run.font
    font.color.rgb = RGBColor(46,188,218)
    run.text = str(cruzamento['ano_formatura'][row])
    


#     #------------------- Minibio -----------------

    # mais uma linha adicionada
    p = tf.add_paragraph()
    p = tf.add_paragraph()
    p = tf.paragraphs[5]
    run = p.add_run()
    font = run.font
    font.bold = True
    font.color.rgb = RGBColor(46,188,218)

    run.text = ""
    run = p.add_run()
    font = run.font
    font.color.rgb = RGBColor(46,188,218)
    run.text = str(cruzamento['minibio'][row])
        
#     p = tf.add_paragraph()
#     p = tf.paragraphs[6]
#     run = p.add_run()
#     font = run.font
#     font.bold = True
#     run.text = "Disponibilidade para Estágio de Férias: "
#     run = p.add_run()
#     run.text = str(cruzamento['Disponibilidade para Estágio de Férias ao fim do Programa'][row])

#     p = tf.add_paragraph()
#     p = tf.paragraphs[7]
#     run = p.add_run()
#     font = run.font
#     font.bold = True
#     run.text = "Disponibilidade para Estágio Quadrimestral: "
#     run = p.add_run()
#     run.text = str(cruzamento['Disponibilidade para Estágio Quadrimestral ao fim do Programa'][row])

#     #------------------- LINKEDIN -----------------

#     p = tf.add_paragraph()
#     p = tf.paragraphs[8]
#     run = p.add_run()
#     font = run.font
#     font.bold = True
#     run.text = "LinkedIn: "

#     run = p.add_run()
#     font = run.font
#     font.size = Pt(12)
#     run.hyperlink.address = cruzamento['Link para o Linkedin'][row]
#     run.text = cruzamento['Link para o Linkedin'][row]

       
#     #------------------- CURRICULO -----------------

    CV_id = CV_ids[row]
    destination_CV = caminho_CV + '//' + str(cruzamento["nome"][row]) + '_CV.pdf'
    CV_path = directory_CV +'//' + str(cruzamento["nome"][row]) + '_CV.pdf'

    try:
        file5 = drive.CreateFile({'id': CV_id})
        file5.GetContentFile(CV_path)

    except:
        pass

    ## download da foto da pessoa
    #------------------- FOTO -----------------

    file_id = f_ids[row]
    destination = caminho_fotos + '//' + str(cruzamento["nome"][row]) + '_foto.jpg'
    img_path = directory_fotos +'//' + str(cruzamento["nome"][row]) + '_foto.jpg'
    file6 = drive.CreateFile({'id': file_id})
    file6.GetContentFile(img_path)
    
    try:
        #add_picture(img_path, left, top, height=height)
        pic = slide.shapes.add_picture(img_path, Cm(3) , Cm(5), height=Cm(12), width=Cm(12))
    except:
        pass


prs.save('Portfolio de Alunos.pptx')
os.startfile('Portfolio de Alunos.pptx')
```
